```python
#Exercise 2
#Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours.

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    overtime = hours - 40
    pay = 40 * rate + overtime * rate * 1.5
else:
    pay = hours * rate

print("Pay:", pay)
```

#Exercise 3
Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

Score           Grade
[ 98, 100]      1.0
[ 94,  97)      1.25
[ 90,  93)      1.5
[ 86,  89)      1.75
[ 82,  85)      2.0
[ 78,  81)      2.25
[ 75,  77)      2.5
[ 0,   74)      5.0

```python
score = float(input("Enter score (0–100): "))

if score < 0 or score > 100:
    print("Invalid score")
elif score >= 98:
    print("Grade: 1.0")
elif score >= 94:
    print("Grade: 1.25")
elif score >= 90:
    print("Grade: 1.5")
elif score >= 86:
    print("Grade: 1.75")
elif score >= 82:
    print("Grade: 2.0")
elif score >= 78:
    print("Grade: 2.25")
elif score >= 75:
    print("Grade: 2.5")
else:
    print("Grade: 5.0")
```
