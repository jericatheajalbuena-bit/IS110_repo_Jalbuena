```python
#Write a program to prompt the user for hours and rate per hour to compute gross pay.
hours_str = input('Enter Hours: ')
hours = float(hours_str)

rate_str = input('Enter Rate: ')
rate = float(rate_str)

gross_pay = hours * rate

print('Pay:', gross_pay)
```

```python
#Assume that we execute the following assignment statements:
#For each of the following expressions, write the value of the expression and the type (of the value of the expression).
width = 17
height = 12.0

result1 = width // 2
print(f"width//2: Value = {result1}, Type = {type(result1)}")

result2 = width / 2.0
print(f"width/2.0: Value = {result2}, Type = {type(result2)}")

result3 = height / 3
print(f"height/3: Value = {result3}, Type = {type(result3)}")

result4 = 1 + 2 * 5
print(f"1 + 2 * 5: Value = {result4}, Type = {type(result4)}")
```

```python
#Write a program that uses input to prompt a user for a number and then outputs the square of the number
num_str = input('Enter a number: ')
num = float(num_str)
square = num * num
print(f'The square of {num} is {square}')
```

```python
#Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.
celsius_str = input('Enter temperature in Celsius: ')
celsius = float(celsius_str)
fahrenheit = (celsius * 9/5) + 32
print(f'The temperature in Fahrenheit is {fahrenheit}')
```
